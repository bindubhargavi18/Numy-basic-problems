```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# 1. Write a Python program to draw a line with suitable label in the x axis, y axis and a title

x = np.array([2020,2021])
y = np.array([100,200])
plt.plot(x,y)
plt.xlabel("years")
plt.ylabel("number of deaths")
plt.title("Corona death rate")
plt.show()
```

```python
# 2. Write a Python program to draw a line using given axis values with suitable label in the x axis , y axis and a title 

id = np.array([1,2,3,4,5])
marks = np.array([100,150,90,200,220])
plt.plot(id, marks)
plt.xlabel("Student id")
plt.ylabel("Marks of students")
plt.show()
```

```python
#3. Write a Python program to draw a line using given axis values taken from a text file,
# with suitable label in the x axis, y axis and a title. 

x = []
y = []
for line in open('test.txt', 'r'):
    line = line.split(' ')
    x.append(line[0])
    y.append(line[1])
plt.plot(x,y)
plt.xlabel("X-Data")
plt.ylabel("Y-Data")
plt.title("Sample Test Data")
plt.show()
```

```python
# 4. Write a Python program to draw line charts of the financial data of Alphabet Inc
# between October 3, 2016 to October 7, 2016. 

df = pd.read_csv("fdata.csv")
x = df.Date
y1 = df.Open
y2 = df.High
y3 = df.Low
y4 = df.Close
plt.plot(x,y1,'b-',label = 'open')
plt.plot(x,y2,'g-',label = 'high')
plt.plot(x,y3,'r-',label = 'low')
plt.plot(x,y4,'k-',label = 'close')
plt.xlabel("Dates")
plt.ylabel("Data")
plt.legend()
plt.grid()
plt.show()
```

```python
# 5. Write a Python program to plot two or more lines on same plot with suitable legends of each line. 

sales1 = np.array([1,6,4,13,17,20])
sales2 = np.array([1,5,2,11,9,7])
months = np.arange(1,7)
# min1 = sales1.min()
# max1 = sales1.max()
# max2 = sales2.max()
plt.plot(months,sales1,'b-')
plt.plot(months,sales2,'g-')
# maxline = plt.plot(max1,max2,'r^')
plt.xlabel("Months")
plt.ylabel("Sales")
plt.legend(['year:2019','year:2020'])
plt.show()
```

```python
#6.  Write a Python program to plot two or more lines with legends, different widths and colors. 

x = np.array([1,2,3,4,5,6])
squares = x*2
cubes = x*3
plt.plot(x,squares,'c',linewidth='5')
plt.plot(x,cubes,'m',linewidth='5')
plt.legend(['squares','cubes'])
plt.title("Squares and cubes")
# plt.grid()
plt.show()
```

```python
#7.  Write a Python program to plot two or more lines with different styles

x = np.linspace(0, 10, 1000)  
plt.plot(x, np.sin(x), '--b', label ='Sine')
plt.plot(x, np.cos(x), 'r-', label ='Cosine')
plt.ylabel("six(x),cos(x)")
plt.xlabel("x")
plt.title("Sine and cosine curves")
plt.legend()
plt.show()
```

```python
#8. Write a Python program to plot two or more lines and set the line markers.

x = np.array([0,30,45,60, 90])
plt.plot(x, np.sin(x), 'o:r', label='sine', ms=10)
plt.plot(x, np.cos(x), 'x:c',label='cosine',ms=10)
plt.legend()
# plt.plot(x, np.tan(x))
plt.show()
```

```python
#9. Write a Python program to display the current axis limits values and set new axis values. 

x=np.array([1,2,3,4,5,7])
y=np.array([2,1,7,4,3,5])
plt.plot(x,y, marker = 'o')
xmin, xmax, ymin, ymax = plt.axis()
s = 'xmin = ' + str(xmin) + ', ' + \
    'xmax = ' + str(xmax) + '\n' + \
    'ymin = ' + str(ymin) + ', ' + \
    'ymax = ' + str(ymax) + ' '
# plt.annotate(s, (4, 6))
print(s)
new_xlim = plt.xlim(0,10)
new_ylim = plt.ylim(0,10)
print("New limits of x axis are:",new_xlim,"\nNew limits of y axis are:",new_ylim)
plt.show()
```

```python
#13. Write a Python program to display the grid and draw line charts of the closing value of Alphabet Inc. 
# between October 3, 2016 to October 7, 2016. Customized the grid lines with linestyle -, width .5. and color blue. 

df = pd.read_csv("fdata.csv")
x = df.Date
y = df.Close
plt.plot(x,y,'r-',label = 'close')
plt.grid(color= 'blue',linestyle='-', linewidth=0.5)
plt.show()
```

```python
#14. Write a Python program to display the grid and draw line charts of the closing value of Alphabet Inc. 
# between October 3, 2016 to October 7, 2016. Customized the grid lines with rendering 
# with a larger grid (major grid) and a smaller grid (minor grid).Turn on the grid but turn off ticks. 

df = pd.read_csv("fdata.csv")
x = df.Date
y = df.Close
plt.plot(x,y,'r-',label = 'close')
plt.grid(which='major', color='blue', linewidth=1.2)
plt.grid(which='minor', color='green', linewidth=0.6)

plt.tick_params(which='both', # Options for both major and minor ticks
                top='off', # turn off top ticks
                left='off', # turn off left ticks
                right='off',  # turn off right ticks
                bottom='off') # turn off bottom tickt.minorticks_on()
# plt.tick_params(which='minor', bottom=False, left=False)
plt.show()
```

```python
# 15. Write a Python program to create multiple plots.

fig = plt.figure()
x = np.arange(0, 10, 0.01)
plt.plot(x, x-5)
  
plt.title("Not so Tiny")
fig.add_subplot(2, 2, 4)
plt.plot(x, 5-x)
  
plt.title("Tiny Plot")
plt.show()
```
