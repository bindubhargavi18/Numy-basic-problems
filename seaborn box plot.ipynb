```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# 2. Write a program to draw a box plot of day by tips for a dataset given in a url 
# https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv 

tips_data = sns.load_dataset('tips')
sns.boxplot(x = 'day',y = 'tip',data = tips_data)
plt.show()
```

```python
# 1. Write a program to draw box plot of life expectancy by continent for a data set given in a url 
# https://raw.githubusercontent.com/resbaz/r-intermediate-gapminder/gh-pages/data/gapminder-FiveYearData.csv

url = "https://raw.githubusercontent.com/resbaz/r-novice-gapminder-files/master/data/gapminder-FiveYearData.csv"
df = pd.read_csv(url)
sns.boxplot(x = df.lifeExp, y = df.continent)
plt.show()
```
