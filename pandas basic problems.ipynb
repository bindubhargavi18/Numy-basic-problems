```python
#2. Write a Python program to convert a Panda module Series to Python list and it's type. 

data_arr = np.array(['dog','cat','rat'])
data_series = pd.Series(data_arr)
print(data_series)
print(type(data_series))
print("coverting series to list:")
data_list = data_series.tolist()
print(data_list)
print(type(data_list))
```

```python
#3.  Write a Python program to add, subtract, multiple and divide two Pandas Series

series1 = pd.Series([2,4,6,8,10])
series2 = pd.Series([1,3,5,7,9])
print("Series1:\n",series1,"\nSeries2:\n",series2)
print("Addition:")
add = series1 + series2
print(add)
print("Subtraction:")
sub = series1 - series2
print(sub)
print("Multiplication:")
mul = series1 * series2
print(mul)
print("Division:")
div = series1 / series2
print(div)
```

```python
#4. Write a Python program to get the powers of an array values element-wise

arr = np.array([0,1,2,3,4,5,6])
arr_series = pd.Series(arr)
print(arr_series)
power = arr_series.pow(arr_series)
print(power)
```
