```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# 1. Write a Python programming to create a pie chart of the popularity of programming Languages. 

pro_lang = ['Java','Python','PHP','JavaScript','C#','C++' ]
popularity = [ 22.2,17.6,8.8,8,7.7,6.7]
fig = plt.figure(figsize =(10, 7))
plt.pie(popularity,labels=pro_lang)
plt.show()
```

```python
# 2. Write a Python programming to create a pie chart with a title of the popularity of programming Languages. 

pro_lang = ['Java','Python','PHP','JavaScript','C#','C++' ]
popularity = [ 22.2,17.6,8.8,8,7.7,6.7]
fig = plt.figure(figsize =(10, 7))
plt.title("Popularity of programming languages")
plt.pie(popularity,labels=pro_lang)
plt.show()
```

```python
# 3. Write a Python programming to create a pie chart with a title of the popularity of programming Languages.
# Make multiple wedges of the pie. 

pro_lang = ['Java','Python','PHP','JavaScript','C#','C++' ]
popularity = [ 22.2,17.6,8.8,8,7.7,6.7]
fig = plt.figure(figsize =(10, 7))
explode = [0, 0.1, 0, 0.2, 0,0.2]
plt.title("Popularity of programming languages")
plt.pie(popularity,labels=pro_lang,autopct='%1.2f%%',explode = explode)
plt.show()
```

```python
# 4. Write a Python programming to create a pie chart of gold medal achievements of five most 
# successful countries in 2016 Summer Olympics. Read the data from a csv file.

df = pd.read_csv('medals.csv')
country = df['country']
gold_medals = df['gold_medal']
fig = plt.figure(figsize =(10, 7))
plt.pie(gold_medals,labels = country,autopct = '%0.2f%%')
plt.show()
```
